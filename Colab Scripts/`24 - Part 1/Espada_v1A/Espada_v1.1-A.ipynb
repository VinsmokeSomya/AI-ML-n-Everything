```python
import tensorflow as tf
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2-large")
```

```python
model = GPT2LMHeadModel.from_pretrained("gpt2-large", pad_token_id=tokenizer.eos_token_id)
```

```python
sentence = 'Who Is Vinsmoke?'
```

```python
input_ids = tokenizer.encode(sentence, return_tensors='pt')
```

```python
input_ids
```

```python
# generate text until the output length (which includes the context length) reaches 200
output = model.generate(input_ids, max_length=200, num_beams=5, no_repeat_ngram_size=2, early_stopping=True)
```

```python
output
```

```python
decoded_output = tokenizer.decode(output[0], skip_special_tokens=True)
```
